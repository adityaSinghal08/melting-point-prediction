```python
#import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load training dataset
df = pd.read_csv('../data/train.csv')
```

```python
#get data info
df.describe().T
```

```python
#identify columns with missing values
missing_cols = [col for col in df.columns if df[col].isnull().sum() > 0]
if missing_cols:
    print("Columns with missing values:", missing_cols)
else:
    print("No columns with missing values.")
```

```python
#identify columns where all values are identical
identical_cols = [col for col in df.columns if df[col].nunique() == 1]
print("Number of columns with identical values:", len(identical_cols))
for col in identical_cols:
    print(f"Column: {col} | Values: {df[col].iloc[0]}")
```

```python
#drop all columns with identical values
df.drop(columns=identical_cols, inplace=True)
```

```python
#identify columns where one value dominates
dominant_cols = []
threshold = 0.99
for col in df.columns:
    top_freq = df[col].value_counts(normalize=True).iloc[0]
    if top_freq >= threshold:
        dominant_cols.append(col)
print("Number of columns with dominant values:", len(dominant_cols))
for col in dominant_cols:
    top_value = df[col].value_counts().idxmax()
    top_freq = df[col].value_counts(normalize=True).iloc[0]
    print(f"Column: {col} | Dominant Value: {top_value} | Frequency: {top_freq:.2f}")
```

```python
#drop all columns with dominant values
df.drop(columns=dominant_cols, inplace=True)
```

```python
#get data info after cleaning
df.info()
```

```python
#find all unique values and count of each value for all columns having group in the name
for col in df.columns:
    if 'group' in col.lower():
        print(f"Column: {col}")
        print(df[col].value_counts())
        print("\n")
```

```python
#save cleaned data as train_improved.csv
df.to_csv('../data/train_improved.csv', index=False)
```
